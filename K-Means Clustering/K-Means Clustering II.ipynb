## K-Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
list(college_data.columns)
```

```python
kmeans.fit(college_data.drop(['Unnamed: 0','Private'],axis=1))
```

```python
kmeans.cluster_centers_
```

## Evaluation
There is no perfect way to evaluate clustering if we don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, we usually won't have this luxury in the real world.

** A new column for college_data called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
college_data['Cluster']=pd.get_dummies(college_data['Private'],drop_first=True) ## creating cluster for evaluation 
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(college_data['Cluster'],kmeans.labels_)
```

```python
print(classification_report(college_data['Cluster'],kmeans.labels_))
```
